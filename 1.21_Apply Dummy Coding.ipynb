```python
# Import Libraries

import pandas as pd

import os
```

#### Read Data

```python
#Location of the file

filepath = os.path.join('Resources', 'titanic_train.csv')
```

```python
#Read and create dataframe

titanic = pd.read_csv(filepath)
titanic.head()
```

---

#### Dummy Coding

```python
# Get the unique values for ``Sex``

titanic.Sex.unique()
```

```python
# Map the values of Sex to a numeric data type
# Store it to a new series in the titanic dataframe


titanic['gender'] = titanic.Sex.map({'male':1, 'female':0})
```

```python
# view the new titanic dataframe with the map-out series gender

titanic.head()
```

---

#### Dummy Code for Embarked using Get Dummies Method

```python
# Get the unique values

titanic.Embarked.unique()
```

```python
# Use the pd.get_dummies method

pd.get_dummies(titanic.Embarked).head()
```

```python
# Store the dummies created for embarked
#use prefix parameter to label the series data to indicate that it is dummified data

embark = pd.get_dummies(titanic.Embarked, prefix='dummy').iloc[:,1:]
```

```python
pd.concat([titanic, embark], axis=1).head()
```
