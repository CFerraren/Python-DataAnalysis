```python
# Import Libraries

import os
import pandas as pd
```

```python
# Declare the file path

filepath = os.path.join('Resources', 'imdb_1000.csv')
```

```python
# Read the csv file into a dataframe using read_csv function

df = pd.read_csv(filepath)
df.head()
```

---

#### Groupby

```python
#Group the data by genre and get the average duration

df.groupby('genre').duration.mean()
```

```python
#Group the data by genre and get the minimum duration

df.groupby('genre').duration.min()
```

```python
#Group the data by genre and get the aggregated functions for duration

df.groupby('genre').duration.agg(['min', 'max', 'median', 'count', 'sum'])
```

```python
#Group the data by genre and applying aggregated functions

df.groupby('genre').agg(['min', 'max', 'median', 'count', 'sum'])
```
