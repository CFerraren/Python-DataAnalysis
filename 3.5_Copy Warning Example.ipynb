```python
# Import Libraries

import pandas as pd
import numpy as np
import os
```

```python
# File Directory Location

file = os.path.join('Resources', 'drinks.csv')
```

```python
# Read data into a dataframe

drinks = pd.read_csv(file)
drinks.head()
```

```python
# Select all the rows with wine_serving equal to 0 and save in a new dataframe

drinks_wine = drinks.loc[drinks.wine_servings==0,:]
drinks_wine
```

```python
# Change the 0 to nan in wine_servings

drinks_wine.loc[:,'wine_servings'] = np.nan
```

```python
# check that although there was copy warning, it was successfull on the nan change

drinks_wine
```

#### To Not Show the Warning

```python
#Use copy method to explicitly copy into a new dataframe

d_wine = drinks.loc[drinks.wine_servings ==0, :].copy()
d_wine.head()
```

```python
d_wine.loc[:, 'wine_servings'] = np.nan
```
